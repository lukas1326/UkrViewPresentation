```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
pop=pd.read_csv("https://population.un.org/wpp/DVD/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2017_TotalPopulationBySex.csv")
```

```python
pop.head()
```

```python
pop.Variant.unique()
```

```python
upop=pop[(pop.Location=='Ukraine')]
var=['Constant fertility','Instant replacement','Zero migration','Constant mortality','No change']
upop=upop.loc[upop.Variant.isin(var)]
upop
```

```python
sns.set()
sns.set_style("whitegrid")
sns.set_context("poster")
g = sns.relplot(x="Time", y="PopTotal", kind="line", data=upop,hue='Variant',aspect=2)
g.fig.autofmt_xdate()
```
