Обновление таблицы Демографічні параметри населення України у 2010-2050 роках.


```python
import pandas as pd
```

```python
pop=pd.read_csv("https://population.un.org/wpp/DVD/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2017_TotalPopulationBySex.csv")
```

```python
pop.head()
```

```python
pop.Variant.unique()
```

```python
years=pop[(pop.Location=='Ukraine')&(pop.Variant=='No change')].Time
poptotal=pop[(pop.Location=='Ukraine')&(pop.Variant=='No change')].PopTotal
Ukrpopulation=pd.DataFrame(index=years,data=poptotal.values,columns=['UkrPop'])

Ukrpopulation.head()
```

```python
Ukrpop=Ukrpopulation[Ukrpopulation.index % 5==0]
Ukrpop.head()
```

```python
#pd.pivot_table(df, values = 'Value', index=['Country','Year'], columns = 'Indicator').reset_index()
Ukrpop=Ukrpop.pivot_table(values='UkrPop',columns='Time')
Ukrpop
```

```python
poptotal=pop[(pop.Location=='World')&(pop.Variant=='No change')].PopTotal
Worldpopulation=pd.DataFrame(index=years,data=poptotal.values,columns=['WorldPop'])
Worldpop=Worldpopulation[Worldpopulation.index % 5==0]
Worldpop=Worldpop.pivot_table(values='WorldPop',columns='Time')
Worldpop
```

```python
allpop=Worldpop.append(Ukrpop)
allpop.to_excel('worldpop.xlsx')
```
