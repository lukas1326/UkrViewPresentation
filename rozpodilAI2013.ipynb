```python
#custom AI group by code
be=pd.read_excel('Сравнение 2013.xlsx',dtype={'Unnamed: 0':str}).replace('nan',)
be=be[['Unnamed: 0','Unnamed: 2','Unnamed: 3']]
be.columns=['kod','name','grupa']
be=be[be.name.notna()]
be=be[be.grupa.notna()]
be.tail(20)
```

```python
grupy=be.grupa.unique()
```

```python
rozpodil=[]
bezpeka_sum=[]
for i in grupy:
    bezpeka=be[be.grupa==i].kod
    bezpeka_sum.append(df.loc[df['1'].isin(bezpeka)][[13]].sum().values) #df.loc[~df['Naimenovanie_klienta'].isin(iskl)]
    rozpodil.append(i)

final=pd.DataFrame({'Rozpodil':rozpodil,'fact2013':bezpeka_sum})
final.fact2013=final.fact2013.apply(lambda x: x[0])
final
```

```python
soz=be[be.grupa=='Социальна підтримка через Мінсоцполитики'].kod
soz1=df.loc[df['1'].isin(soz)]
sozplus=['2500000','2510000']
sozminus=['2506020','2506050','3501160']
dif=df.loc[df['1'].isin(sozplus)][13].sum()-df.loc[df['1'].isin(sozminus)][13].sum()
dif
```

```python
fin=be[be.grupa=='Фінансі'].kod
finplus=['3500000','3510000']
finall=df.loc[df['1'].isin(fin)]
finminus=finall.loc[~finall['1'].isin(finplus)]
diffin=df.loc[df['1'].isin(finplus)][13].sum()-finminus[13].sum()
diffin
```

```python

final.loc[final.Rozpodil=='Социальна підтримка через Мінсоцполитики','fact2013']=dif
final.loc[final.Rozpodil=='Фінансі','fact2013']=diffin
#final.to_excel('udalit.xlsx')
final
```
