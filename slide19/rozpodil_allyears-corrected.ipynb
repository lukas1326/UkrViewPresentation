```python
#custom AI group by code
be=pd.read_excel('data/rozpodil.xlsx',dtype={'Unnamed: 0':str}).replace('nan',)
be=be[['Unnamed: 0','Unnamed: 2','Unnamed: 3']]
be.columns=['kod','name','grupa']
be=be[be.name.notna()]
be=be[be.grupa.notna()]
```

```python
be
# создается кастомная группировка по примеру готового файла за 2013год,логика неизвестна
```

```python
grupy=be.grupa.unique()
grupy=grupy.tolist()
# создается уникальный список групп, в которых суммируются значения
# исключаются группы "Фінанси" и "Социальна підтримка через Мінсоцполитики"
# потому что там уникальная логика суммирования и вычитания статей
del grupy[9]
del grupy[-4]
grupy
```

```python
years=[df2013,df2014,df2015,df2016,df2017,df2018]
```

```python
 def sum_result(x):
    all_sum=[]
    for i in years:
        cody=be[be.grupa==grupy[x]].kod.astype(str)
        table=i.loc[i.code.isin(cody)]
        all_sum.append(table.all_done.sum())
        
    return all_sum  
# возвращает сумму по определенной группе и по годам 
```

```python
z=[]
for i in range(len(grupy)):
    z.append(sum_result(i))
# собираем значения сумм
```

```python
df=pd.DataFrame(z,index=grupy,columns=['2013','2014','2015','2016','2017','2018'])
df
# формируем датафрейм
```

```python
# частный случай с исключенными группами - вычитаем некоторые статьи
soz=[]
sozplus=['2500000','2510000']
sozminus=['2506020','2506050','3501160','2507000']
for i in years:
    dif=i.loc[i.code.isin(sozplus)]['all_done'].sum()-i.loc[i.code.isin(sozminus)]['all_done'].sum()
    soz.append(dif)

soz
```

```python
df_soz=pd.DataFrame([soz],index=['Социальна підтримка через Мінсоцполитики'],columns=['2013','2014','2015','2016','2017','2018'])
df_soz
df=df.append(df_soz)
# добавляем полученные строки в датафрейм
```

```python
diffin=[]
for i in years:
    fin=be[be.grupa=='Фінансі'].kod
    finplus=['3500000','3510000']
    finall=i.loc[i.code.isin(fin)]
    finminus=finall.loc[~finall.code.isin(finplus)]
    diffin.append(i.loc[i.code.isin(finplus)]['all_done'].sum()-finminus['all_done'].sum())
diffin
```

```python
df_fin=pd.DataFrame([diffin],index=['Фінанси'],columns=['2013','2014','2015','2016','2017','2018'])
df=df.append(df_fin)
df
```

```python
# по некоторым группам меняются номера статей, но название остается
# поэтому корректируем вручную
#так как добавляем "внутренние" статьи,их нужно вычесть из других групп
#чтобы не повлияло на общую сумму
#общая сумма равна сумме по статьям,которые оканчиваются на "0000"
df2014[df2014.name=='Обслуговування державного боргу'].all_done
df.iloc[1]['2014']=df2014[df2014.name=='Обслуговування державного боргу'].all_done
df.loc['Фінанси','2014']=df.loc['Фінанси','2014']-df.iloc[1]['2014']
```

```python
df.iloc[1]['2015']=df2015[df2015.name=='Обслуговування державного боргу'].all_done
df.loc['Фінанси','2015']=df.loc['Фінанси','2015']-df.iloc[1]['2015']
```

```python
df.iloc[1]['2016']=df2016[df2016.code=='3511350'].all_done#была статья 3501170,в 2016 стала 3511350,вычитается из Финансов
df.loc['Фінанси','2016']=df.loc['Фінанси','2016']-df.iloc[1]['2016']
df.iloc[9]['2016']=df2016[df2016.code=='2506080'].all_done
df.loc['Социальна підтримка через Мінсоцполитики','2016']=df.loc['Социальна підтримка через Мінсоцполитики','2016']-df.iloc[9]['2016']
```

```python
df.iloc[1]['2017']=df2017[df2017.code=='3511350'].all_done
df.loc['Фінанси','2017']=df.loc['Фінанси','2017']-df.iloc[1]['2017']
df.iloc[9]['2017']=df2017[df2017.code=='2506080'].all_done
df.loc['Социальна підтримка через Мінсоцполитики','2017']=df.loc['Социальна підтримка через Мінсоцполитики','2017']-df.iloc[9]['2017']
df.iloc[1]['2018']=df2018[df2018.code=='3511350'].all_done
df.loc['Фінанси','2018']=df.loc['Фінанси','2018']-df.iloc[1]['2018']
df.iloc[9]['2018']=df2018[df2018.code=='2506080'].all_done
df.loc['Социальна підтримка через Мінсоцполитики','2018']=df.loc['Социальна підтримка через Мінсоцполитики','2018']-df.iloc[9]['2018']
df
# итоговая таблица
```
